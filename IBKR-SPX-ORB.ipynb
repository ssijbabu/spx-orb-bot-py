### Connect to IB
- Specify port/client
- util.startLoop() is needed for Jupyter notebook: BTS it does nest_asyncio.apply()

```python
from ib_async import *
util.startLoop()
ib = IB()
ib.connect(port=4004, clientId=1)
```

```python
contract = Contract(symbol='SPX', secType='IND', exchange='CBOE', currency='USD')
ib.qualifyContracts(contract)
```

```python
index = Index('SPX', 'CBOE', 'USD')
ib.qualifyContracts(index)
```

```python
short_call_option = Option('SPX', '20251114', 6630, 'P', 'SMART')
long_call_option = Option('SPX', '20251114', 6645, 'P', 'SMART')
ib.qualifyContracts(short_call_option, long_call_option)
```

```python
import pandas as pd
```

```python
chains = ib.reqSecDefOptParams(index.symbol, '', index.secType, index.conId)
chain = [c for c in chains if c.tradingClass == 'SPX' and c.exchange == 'SMART']
util.df(chain).explode('expirations').explode('strikes')
```
